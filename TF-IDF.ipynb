Importing libraries

```python
import texthero as hero
import pandas as pd
import nltk
import itertools
nltk.download('inaugural')
from nltk.corpus import inaugural

from sklearn.feature_extraction.text import TfidfVectorizer
```

Text documents

```python
doc1=(inaugural.words("/content/drive/MyDrive/Colab Notebooks/NLP/doc1.txt"))
doc2=(inaugural.words("/content/drive/MyDrive/Colab Notebooks/NLP/doc2.txt"))
doc3=(inaugural.words("/content/drive/MyDrive/Colab Notebooks/NLP/doc3.txt"))
doc4=(inaugural.words("/content/drive/MyDrive/Colab Notebooks/NLP/doc4.txt"))
```

Text processing

```python
corpus = [doc1,doc3,doc4]
for i in range(3):
    corpus[i]=pd.Series(corpus[i])
    corpus[i]=hero.remove_punctuation(corpus[i])   #Remove all string.punctuation (!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~).
    corpus[i]=hero.remove_diacritics(corpus[i])    #Remove all accentuated characters from strings.
    corpus[i]=hero.remove_stopwords(corpus[i])      #Remove all stop words.
    corpus[i]=hero.remove_whitespace(corpus[i])    #Remove whitespace
    corpus[i]=hero.lowercase(corpus[i])              #Lowercase all text.
    corpus[i]=hero.remove_digits(corpus[i])         #Remove all blocks of digits.
```

Finding 50 words with high word counts

```python
both = nltk.FreqDist(corpus[0]+corpus[1]+corpus[2])
both_most_common = both.most_common()
words=list(itertools.chain(*(sorted(ys) for k, ys in itertools.groupby(both_most_common, key=lambda t: t[1]))))
top_50_words=[]
for i in range(50):
    top_50_words.append(words[i][0])
print(top_50_words)
```

TF-IDF

```python
vec=TfidfVectorizer()
for i in range(3):
  corpus[i]=str(corpus[i])
matrix=vec.fit_transform(corpus)
print("Feature Names n",vec.get_feature_names_out())
print("Sparse Matrix n",matrix.shape,"n",matrix.toarray())
print(matrix)
```
