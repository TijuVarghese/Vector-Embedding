```python
import texthero as hero
import pandas as pd
import nltk
import itertools
from nltk.corpus import inaugural
from gensim.models import KeyedVectors
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.lm import Vocabulary
import gensim
from gensim.models import Word2Vec
```

```python
print("Number of words",len(inaugural.words("/content/drive/MyDrive/Colab Notebooks/mughal.txt")))
corpus=(inaugural.words("/content/drive/MyDrive/Colab Notebooks/mughal.txt"))
print(corpus)
```

```python
corpus=pd.Series(corpus)
```

Text processing

```python
corpus=hero.remove_punctuation(corpus)   #Remove all string.punctuation (!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~).
corpus=hero.remove_diacritics(corpus)    #Remove all accentuated characters from strings.
corpus=hero.remove_stopwords(corpus)      #Remove all stop words.
corpus=hero.remove_whitespace(corpus)    #Remove whitespace
corpus=hero.lowercase(corpus)              #Lowercase all text.
corpus=hero.remove_digits(corpus)         #Remove all blocks of digits.
```

```python
fd = nltk.FreqDist(corpus)
print(fd)
fd_1=fd.most_common()
list(itertools.chain(*(sorted(ys) for k, ys in itertools.groupby(fd_1,key=lambda t: t[0]))))
```

Top 50 most occurring words

```python
words=list(itertools.chain(*(sorted(ys) for k, ys in itertools.groupby(fd_1,key=lambda t: t[0]))))
top_50_words=[]
for i in range(49):
    top_50_words.append(words[i][0])
print(top_50_words)
```

CBOW -Word2vec

```python
model = gensim.models.Word2Vec(corpus, min_count = 1,size = 50, window = 5, sg=0) 
model.train([top_50_words],total_examples=1, epochs=1)
model.save('CBOW.embedding')
new_model = gensim.models.Word2Vec.load('CBOW.embedding')
```
